```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```
%matplotlib inline
```

$$
R^{2}=1-\frac{\Sigma(y-\hat{y})^{2}}{\Sigma(y-\bar{y})^{2}}
$$


```
def compute_r2_numpy(y_true, y_predicted):
    Nr = np.sum(np.square(y_true - y_predicted))
    Dr = np.sum(np.square(y_true - np.mean(y_true)))
    R_squared = 1 - (Nr/Dr)
    return R_squared
```

```
x = np.arange(4)
y_true = np.arange(4)
```

```
plt.scatter (x, y_true)
plt.show()
```

```
y_pred = np.array([0, 1, 2, 3])
```

```
plt.scatter (x, y_true, s=50, color='red')
plt.plot(x, y_pred)
plt.show()
```

```
compute_r2_numpy(y_true, y_pred)
```

```
r2_score(y_true, y_pred)
```

```
y_pred = np.array([3, 2, 1, 0])
```

```
plt.scatter (x, y_true, s=50, color='red')
plt.plot(x, y_pred)
plt.show()
```

```
compute_r2_numpy(y_true, y_pred)
```

```
r2_score(y_true, y_pred)
```
